**Distribution Plots**

Below are some different ways to plot distribution plot

```python
sns.distplot(crash_df['not_distracted'])
```

It gives a histogram plot with a line called kernel density estimation(kde). To remove it pass kde as False

```python
sns.distplot(crash_df['not_distracted'], kde = False)
```

**Joint Plot**
Compares two distribution and plots a scatter plot by default

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'kde')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'reg')
```

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'hex')
```

**KDE Plot**

```python
sns.kdeplot(crash_df['alcohol'])
```

**PairPlot**
Plots relationship across entire dataframe in numerical values

```python
sns.pairplot(crash_df)
```

**RugPlot**:

Plot single column of datapoints in a dataframe as sticks.The sticks are denser where the values are common unlike histogram where the plot is taller where there are more values

```python
sns.rugplot(crash_df['total'])
```

**Categorical Plots**

Below are some different ways to plot categorical plot

**Bar Plot**

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.info()
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = tips_df, estimator = np.cov)    #estimator is set at the variance between the axes and the data by default is aggregated as the mean but can be changed using different parameters. Can pass user defined func also
```

**CountPlot**

Same as Barplot but estimator counts the number of occurances

```python
sns.countplot(x = 'sex', data = tips_df)
```

Hue parameter:- It determines which column in the data frame should be used for color encoding

Hue provides an additional way for us to add in another category

```python
sns.countplot(x = 'sex',hue= 'smoker',data = tips_df)
```

**BoxPlot**

Compare different variables and show quartiles of data

boxes show quartiles of data

line in the box show median

box is going to extend one standard deviation from median

wiskers: line from the box extends to all the data except the ouliers or deviation of current data

points in the plot are the outliers

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')

plt.legend(loc= 0)
```

**ViolinPlot**

It is a combination of box plot as well as kde plot

Box plot corresponds to datapoints while violin plot uses kde estimation of datapoints

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')

plt.legend(loc = 0)
```

Can use split which compares category to each other

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', split = 'True')

plt.legend(loc = 0)
```

**StripPlot**

It draws a scatter plot representing all the datapoints where one variable is categorical.

They are often used to show different observations with a box plot that is goig to represent average distributions

```python
plt.figure(figsize = (8, 5))

sns.stripplot(x = 'day', y = 'total_bill', data = tips_df)
```

Can use jitter attribute that is going to spread our dataset out so that they are not stacked on top of each other

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True)
```

When we use dodge with hue it will separate the categories of hue 

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, jitter = True, hue = 'sex', dodge = 'True')
```

**Swarm Plot**

Its like a strip plot and its adjusted in a way that points dont overlap 

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df)

sns.swarmplot(x = 'day', y = 'total_bill', data = tips_df, color= 'white')
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips_df)
```

```python
plt.figure(figsize = (8, 5))

sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex')
```

Palette defines different color schemes for the plot

```python
sns.set_style('white')

sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', palette='magma')
```

To place legend other than given by loc = 0 we can either increase size of the figure or shift it to different corners 

upper right: 1

upper left: 2

lower left: 3

lower right: 4

**MatrixPlot**

**HeatMaps**

In heatmap we form a matrix using column of dataset as both the axes

Its is done by two ways:

1) setup a corelation which tells how influential a variable can be on result

2) use a pivot table to form a matrix

**Corelation**

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)
crash_df['abbrev'] = pd.to_numeric(crash_df['abbrev'], errors='coerce')

crash_matrix = crash_df.corr()

sns.heatmap(crash_matrix, annot= True, cmap = 'Blues')          #annotation defines the number of influence
```

**Pivot Table**

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(values = 'passengers', index = 'month', columns = 'year')

sns.heatmap(flights, cmap = 'Blues', linecolor='white', linewidths=1)
```

**ClusterMaps**

Its like clustering similar data points.

It gives hierrachially clustered heat maps. In this distance between points is calculated and the closset are joined and this continues for next closet and then compare columns and the rows of the heatmap

Heatmap does not reposition data to find clusters of data while clustermap is going to tightly fit 
and cluster data to see specific patterns.

```python
iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
```

standard_scale normalize the data to focus on clustering

```python
sns.clustermap(flights, cmap = 'Blues', standard_scale = 1)
```

**Pairgrid**

In pair plot we get different kinds of plots but if we want to have specific kind of plot for specific data pairgrid is uesd

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue = 'species')

iris_g.map(plt.scatter)
```

Now if we want to have scatter plot for different plots upper lower and the diagonal and also a histogram across diagonal


```python
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

We want to have scatter plot show off of the diagonal in the upper and the lower part

upper is the upper part of the diagonal and  lower is the lower part of the diagonal

```python
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

We plot a scatter plot on the upper part and kde plot on the lower part

```python
iris_g = sns.PairGrid(iris, hue = 'species')
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
```

We can also define x and y for our custom grid

```python
iris_g = sns.PairGrid(iris, hue = 'species', x_vars = ['sepal_length', 'sepal_width'], y_vars = ['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

**FacetGrid**

We can plot multiple plots in a grid in which we can define our columns and rows

Say we want to plot histogram for smokers and non-smokers with total bill for lunch and dinner

```python
tips_fg = sns.FacetGrid(tips_df, col = 'time', row = 'smoker')
```

```python
tips_fg = sns.FacetGrid(tips_df, col = 'time', row = 'smoker')
tips_fg.map(plt.hist, 'total_bill', bins = 8)
```

```python
tips_fg = sns.FacetGrid(tips_df, col = 'time', row = 'smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

Assign colors to different variables

aspect is the width based on height if aspect is 1 then ratio is equal 

```python
tips_fg = sns.FacetGrid(tips_df, col = 'time', hue = 'smoker', height = 4, aspect = 1.3, col_order = ['Dinner', 'Lunch'], palette= 'Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor= 'white')
```

```python
kws = dict(s = 50, linewidth = 0.5, edgecolor = 'w')        #s- size of markers

tips_fg = sns.FacetGrid(tips_df, col = 'sex', hue = 'smoker', height = 4, aspect = 1.3,
                         hue_order = ['Yes', 'No'], hue_kws = dict(marker = ['^', 'v']))

tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

We want to put each person in their own plot with five per line and plot each of their score depending upon how much attention they were able to pay during the testing phase

```python
att_df = sns.load_dataset('attention')

att_fg = sns.FacetGrid(att_df, col = 'subject', col_wrap = 5, height = 1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker = '.')
```
